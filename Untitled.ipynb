## Statistical Test

Null Hypothesis (H): There is no significant association between the event_type and whether a purchase is made or not (i.e., event type and purchase are independent).

Alternative Hypothesis (H): There is a significant association between the event_type and whether a purchase is made or not (i.e., event type and purchase are dependent).

```python
df.head()
```

```python
df.info()
```

```python
df['event_type'].unique()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming 'df' is your DataFrame containing 'event_type' and 'purchase_made'
# You can create the 'purchase_made' variable from 'event_type', for example:
df['purchase_made'] = df['event_type'].apply(lambda x: 1 if x == 'purchase' else 0)

# Create the contingency table
contingency_table = pd.crosstab(df['event_type'], df['purchase_made'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Check the p-value
print(f"P-value: {p}")
if p < 0.05:
    print("Reject the null hypothesis: There is a significant relationship between event_type and purchase_made.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between event_type and purchase_made.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming your DataFrame is named 'df'
# Step 1: Create a contingency table
contingency_table = pd.crosstab(df['event_type'], df['product_id'])

# Step 2: Perform the Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)

if p < 0.05:
    print("Reject the null hypothesis: There is a significant relationship between event_type and product_id.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between event_type and product_id.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming 'df' is your DataFrame
# Create a contingency table
contingency_table = pd.crosstab(df['event_type'], df['category_id'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Chi-Square Test Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)

if p < 0.05:
    print("Reject the null hypothesis: There is a significant relationship between event_type and category_id.")
else:
    print("Fail to reject the null hypothesis: No significant relationship between event_type and category_id.")
```
